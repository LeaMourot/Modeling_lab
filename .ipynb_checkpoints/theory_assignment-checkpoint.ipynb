# 1. Loss functions

Lets focus on linear regression of the form 

$\mathbf{y} \approx f(\mathbf{X}) = \mathbf{X}\mathbf{w_1} + \mathbf{w_0}.$


#### 1.1 What are the rows of $\mathbf{X}$?

The rows of ${X}$ corresponds to the ${m}$ observations that describe the system

#### 1.2 What are the columns of $\mathbf{X}$?

The columns of ${X}$ corresponds to the ${n}$ features that describe the system.

Often, we write the equation above as

$\mathbf{y} \approx \mathbf{\tilde{X}}\mathbf{w}$


#### 1.3 How does $\mathbf{\tilde{X}}$ look like in this case (i.e., how does the shape of the matrix change compared to $\mathbf{X}$)? (1 point)


In this last equation, the bias term $\mathbf{w_0}.$ is incorporated into the the $\mathbf{\tilde{X}}$ matrix. As a consequence, compared to the $\mathbf{X}$ matrix, this $\mathbf{\tilde{X}}$ matrix will always shows a supplementary column and this no matter the total column number of $\mathbf{X}$.

For machine learning, we need a cost function. Two common choices are the mean-squared error (MSE, $\mathcal{L}_2$), and the mean-absolute error (MAE, $\mathcal{L}_1$)

\begin{align}
    \mathcal{L}_2 &=& \frac{1}{N} \sum_{i=1}^N \left(y_i - f(x_i) \right)^2 \\
    \mathcal{L}_1 &=& \frac{1}{N} \sum_{i=1}^N \left|y_i - f(x_i) \right| 
\end{align}

#### 1.5 In the Jupyter notebook, write a Python function that computes these two cost functions given an error term $\boldsymbol{\epsilon} = \mathbf{y} - \mathbf{\tilde{X}}\mathbf{w}$ (2 points)

````python
def mean_squared_error(error_vector):
    N = len(error_vector)
    squared_error_vector = [i**2 for i in error_vector]
    mean_squared_error = 1/N *sum(squared_error_vector)
    return mean_squared_error
````

````python
def mean_absolute_error(error_vector):
    N = len(error_vector)
    absolute_error_vector = [abs(i) for i in error_vector]
    mean_absolute_error = 1/N*sum(absolute_error_vector)
    return mean_absolute_error
````

#### 1.6 What is the shape of these cost functions as a function of the error  (1 point)

````python
x_axis = np.linspace(-10,10,100) # change as you wish for your plot
x_axis_list = [[float(x)] for x in x_axis]

y_mae = [mean_absolute_error(x) for x in x_axis_list]
y_mse = [mean_squared_error(x) for x in x_axis_list]
````

````python
plt.plot(x_axis, y_mse, label='MSE')
plt.plot(x_axis, y_mae, label='MAE')
plt.xlabel('error term')
plt.ylabel('cost function')
plt.legend()
````

#### 1.7  Are both loss functions differentiable for all $\boldsymbol{\epsilon}$? (1 point) What implications does this have for gradient based optimization like gradient descent? (1 point)

While the mean square error function is differentiable for all $\boldsymbol{\epsilon}$, the mean absolute error function is a bit different. Indeed, this last function is not differentiable at $\boldsymbol{\epsilon}$ __= 0__ point due to the presence of the absolute value term in its formula. This difference between the two functions can be seen on the graph plotted above (question 1.6). On this plot, one can observe that while the MSE shows a smooth transition from negative to positive slope, the MAE function slope transition is really sharp.
Finally, this non differentiability of the MAE function will prevent the use of gradiant based optimization methods since this function gradiant is undefined when $\boldsymbol{\epsilon}$ __= 0__.

#### 1.8 Which loss function is more sensitive to outliers (1 point) and why (1 point)?

The mean square error function is more sensitive to outliers since it will amplify quadratically the given errors. As a consequence, large errors will stand out much more prominently in comparison with the rest of the set.

# 2. Regularization

Assume that the columns of $\mathbf{X}$ are linearly independent.
As a refresher of linear algebra, recall when the linear system $\mathbf{X}\mathbf{w} = \mathbf{y}$ has

#### 2.1 One unique solution (1 point)

The system ${Xw}= {y}$ has a unique solution when, for a matrix ${X}$  with linearly independent columns : ${Rank(X)} = {Rank(X|y)}$ where ${X|y}$ is the augmented matrix concatenating ${X}$  and ${y}$ . This is equivalent to admitting that our scaled and reduced matrix ${X}$  must have one pivot per column. 

#### 2.2 No solution (1 point)

This system has no solution when : ${Rank(X)}<{Rank(X|y)}$. In fact, having linearly independent columns, the reduced scaled matrix ${X}$  will have one pivot per column, but in the case the previous condition isn't fullfilled, the system will not be solvable anymore since one of the ${X|y}$  matrix row will state an impossible equation such as ${0}={C}^{ste}$.

#### 2.3 An infinite number of solutions (1 point)

With linearly independent columns, there are no infinite solutions. A matrix with linearly independent columns admits at most one solution for this system.

To calculate the weights, $\mathbf{w}$ we have to solve
\begin{equation}
    \mathbf{w} = \left(\mathbf{\tilde{X}}\mathbf{\tilde{X}}^T\right)^{-1}\mathbf{\tilde{X}}^T\mathbf{y}
\end{equation}

#### 2.5 In general, why can't we solve eq.~\ref{eq:linear_reg} using $\mathbf{y} = \tilde{\mathbf{X}}^{-1}\textbf{w}$? (1 point)

In general, we can't solve this equation using the tool above because there might be issues with the X matrix invertibility. Indeed, only square matrices are inversible. In the case where X is square, there would be no problem using its inverse vesrion ${X}^{-1}$ to determine w. However, as in most cases, ${X}$ isn't square so its inverse matrix ${X}^{-1}$ doesn't exist. As a consequence, another method should be used in order to solve the system.

#### 2.5  What happens if some columns are linearly dependent? (1 point) What is the connection to feature selection? (1 point)

If some columns are now linearly dependent and the ${XX}^{T}$ matrix is no longer reversible, the formule obtained earlier cannot be used to calculate the weigths anymore.

As its name suggests,the feature selection enables to select the most important and independent characteristics of the system so that, when put into matrix form, our data corresponds to a matrix with linearly independent columns. In this way, we will no longer have matrices inversion problems to perform the different calculations.

#### 2.6  What is the shape of the parabola as a function of $a$?

````python
def parabola(x, a = 1): 
    return a * x ** 2

x_axis_parabola = np.linspace(-10, 10, 100)
a_values = [5, -1, 1]
colors = ['blue', 'red', 'green']
labels = ['a = 5', 'a = -1', 'a = 1']


plt.figure(figsize=(10, 6))


for a, color, label in zip(a_values, colors, labels):
    plt.plot(x_axis_parabola, parabola(x_axis_parabola, a), label=label, color=color)

plt.title('Tracé des paraboles pour différentes valeurs de a')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.ylim(-10, 10)  
plt.xlim(-10, 10)
plt.show()
````

The orientation of the parabola will change according to the sign of ${a}$. If ${a}$ is positive, the parabola faces upwards whereas for ${a}$ negative, the parabola faces downwards. Moreover, the absolute value of ${a}$ will impact the shape of the parabola : the greater the ${a}$ value, the sharper the curve's slope as ${x}$ moves away from zero.

#### 2.7 Plot the approximation to the function for different order polynomials ($N \in \{1, 2, 16\}$) and with different regularization strength ($\lambda \in \{0, 10^{-3}, 10^{-2}, 1\}$). What do you observe 

````python
def true_function(X):
    return np.cos(1.5 * np.pi * X)
````

````python
X_test = np.linspace(0, 1, 100) # some grid for us on the x axis
````

````python
n_samples = 10 # the number of points we will sample from true_function
degrees = [1, 2, 16] # the polynomial degrees we will test

X = np.sort(np.random.rand(n_samples))
y = true_function(X) + np.random.randn(n_samples) * 0.1 # add some scaled random noise
````

````python
for i in range(len(degrees)):
    degree = degrees[i]
    alphas = [0, 0.001, 0.01, 1]
    
    # Plot the noisy data and the true function
    plt.scatter(X, y, label="Noisy samples", color="blue")
    plt.plot(X_test, true_function(X_test), label="True function", color="black", linewidth=2)
    for alpha in alphas:
        model = Pipeline([
            ("polynomial_features", PolynomialFeatures(degree=degree, include_bias=False)),
            ("ridge_regression", Ridge(alpha=alpha))
        ])
    
        # Fit the model and predict
        model.fit(X[:, np.newaxis], y)
        y_pred = model.predict(X_test[:, np.newaxis])
    
        # Plot the result for this alpha
        label = f"Alpha {alpha}"
        plt.plot(X_test, y_pred, label=label)
    
    # Add plot labels and legend
    plt.title(f"Polynomial Approximation (Degree {degree}) with Different Regularization")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()
````

Here, we notice a problem regularizing the function to a degree of 16. This is probably due to the very large degree of the function one would like to use in order to approximate our true function. To study the regularizing effect at a 16 polyomial degree, a different plot will be computed for each alpha value progressively from 0 to 1.

````python
for alpha in alphas:
    plt.figure()

    #regression
    polynomial_features = PolynomialFeatures(degree=16, include_bias=False)
    ridge_regression = Ridge(alpha=alpha)
    pipeline_ridge = Pipeline([
        ("polynomial_features", polynomial_features),
        ("ridge_regression", ridge_regression)
    ])
    pipeline_ridge.fit(X[:, np.newaxis], y)

    # Plot
    plt.plot(X_test, pipeline_ridge.predict(X_test[:, np.newaxis]), 
             label=f"Model of degree 16 with alpha = {alpha}")
    plt.scatter(X, y, label='Noisy samples')
    plt.plot(X_test, true_function(X_test), c='r', label='True function')
    plt.title(f"Polynomial Approximation (Degree 16) with Regularization alpha = {alpha}")
    plt.legend()
    plt.xlabel('X-axis label')  
    plt.ylabel('Y-axis label')   
    plt.show()
````

Comparing these different plots, one can determine the best polynomial order and regularization strength value for this true function. Firstly, the degree-one and degree-two models are clearly not sufficient for the true function approximation since their shapes do no match the true function's one no matter the regularization strength value. The highest polynomial degree model (N = 16) is much more complex but constitute the best true function approximation among the different dregree model tested. Indeed, for some alpha values its shape is really close to the true function's shape.


Finally if we look at the different aplha values for this last polynomial degree model, one can see that 0.001 and 0.01 alpha values models constitute the most accurate approximation of the true function among the different propositions. Indeed, a regularization strength value of 1 is too high : a significant amount of informations is neglected leading to an underfitting of the true function. Conversely, a regularisation strength value of 0 is too low : too much data complexity is captured leading to an overfitting the true function. The complexity at this alpha value is so high that this time again the programme cannot properly plot the model.


As a conclusion the higher the regularization strength, the lower the complexity of the model. Choosing lower apha values makes the model simpler but might not approximate the true function well enough. As far as polynomial degree is concerned in this study, the higher the degree, the greater the model approximation. Among the different polynomial degrees and regularization strength values proposed, the best approximation model is given for alpha = 0.01 or 0.001 at N = 16. Nevertheless, a better approximation could be found by testing more polynomial orders (maybe some higher than 16) and regularization strenght (still between 0 and 1).

#### 2.8 What do you observe if you change the number of samples from the function?

In this question, the true function the regularization results for a number of samples of 10 and 100 are compared. An N=16 polynomial degree and a alpha=0.01 regularization strength are kept since they show the best approximation model of the given true function (as detailed in the previous part).

````python
sample_sizes = [10, 100]  
degree = 16  
alpha = 0.01  
for n_samples in sample_sizes:
   
    X = np.sort(np.random.rand(n_samples))
    y = true_function(X) + np.random.randn(n_samples) * 0.1  

   
    plt.scatter(X, y, label=f'{n_samples} noisy samples')
    plt.plot(X_test, true_function(X_test), c='k', label='true function',linewidth=2)

    polynomial_features = PolynomialFeatures(degree=degree, include_bias=False)
    ridge_regression = Ridge(alpha=alpha)
    pipeline_ridge = Pipeline([
        ("polynomial_features", polynomial_features),
        ("ridge_regression", ridge_regression)
    ])
    
    pipeline_ridge.fit(X[:, np.newaxis], y)
    
    
    plt.plot(X_test, pipeline_ridge.predict(X_test[:, np.newaxis]), 
         label=f"Model (N={n_samples}, Degree={degree}, Alpha={alpha})", c='red')
    plt.title("Polynomial Regression with Ridge Regularization")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()
````

Thanks to these two plots, one can see that the model's shape is much closer to the true function for a sample number of 100 than for a sample number of 10. Therefore, increasing the number of samples will give a better approximation to the true function.
