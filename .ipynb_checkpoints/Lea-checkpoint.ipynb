# PART C : Data plots

In this part we choose to represent how the density of an atom changes acording to its atomic mass through a 2 dimensional scatter plot. For the second plot, we decided to show the proportions of each atom types in the Periodic Table into an pie chart.

### Group block pie chart

```python
#Number of element in each group block
noble_gas = df['GroupBlock'].value_counts().get('Noble gas', 0)
non_metal= df['GroupBlock'].value_counts().get('Non metal', 0)
alkali_metal= df['GroupBlock'].value_counts().get('Alkali metal', 0)
alkali_earth_metal= df['GroupBlock'].value_counts().get('Alkaline earth metal', 0)
metalloid= df['GroupBlock'].value_counts().get('Metalloid', 0)
halogen= df['GroupBlock'].value_counts().get('Halogen', 0)
post_tansition_metal= df['GroupBlock'].value_counts().get('Post-transition metal', 0)
transition_metal= df['GroupBlock'].value_counts().get('Transition metal', 0)
Lanthanide= df['GroupBlock'].value_counts().get('Lanthanide', 0)
Actinide= df['GroupBlock'].value_counts().get('Actinide', 0)

#Lists
Groupblocks = ['Noble gas','Non metal','Alkali metal','Alkaline earth metal','Metalloid','Halogen','Post-transition metal','Transition metal','Lanthanide','Actinide']
Groupblocks_types = [noble_gas, non_metal, alkali_metal, alkali_earth_metal, metalloid, halogen, post_tansition_metal, transition_metal, Lanthanide, Actinide]
Proportions = [type/118 for type in Groupblocks_types]

#pie chart
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(Proportions, autopct='%1.1f%%', startangle=140)
plt.legend(wedges, Groupblocks, title="GroupBlock Types", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.title('Proportions of GroupBlock Types')
plt.axis('equal') 
plt.show()
    
```

From this plot, one can see that the elements of the periodic table are not equally distributed into the different group  blocks. The transition metals forms the biggest group block containing 34.2% of the periodic table's elements. In contrast, the proportion of elments contained by the Non metal group bock is so small that 0% is displayed on the chart.

### Atomic mass/Density scatter plot

```python
#lists
Atomic_mass= df['AtomicMass'].tolist()
AtomicNumber = df['AtomicNumber'].tolist()

#plots
plt.figure(figsize=(10, 6))
plt.scatter(AtomicNumber, Atomic_mass, alpha=0.7)
plt.xlabel('AtomicNumber')
plt.ylabel('Atomic_mass')
plt.grid()
plt.show()
```

On this plot, one can see that the atomic number is proportional to the atomic mass since the curve obtained is linear. This result was expected. Indeed, the atomic number is an indication of the number of electrons that composes the atom : the larger the atomic number, the more the electrons, the heavier the atom (higher atomic mass).
